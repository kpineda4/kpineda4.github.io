```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
opiods = pd.read_csv('OpidsVA.csv')
```

```python
opiods
```

```python
#Do opioid overdoes tend to be associated with less affluent areas—that is, areas where families have lower incomes?
#Scatterplot shows relationships between two variables. Clearly, we can see that there is no relationship between
#income and opiod overdoses. 

fig, ax = plt.subplots()
ax.scatter(opiods['MedianHouseholdIncome'], opiods['FPOO-Rate'], alpha=0.5, c='b')
fig.suptitle('Median Household Income vs. Opiod Overdose Rate in VA')
ax.xaxis.set_label_text('Median HH Income')
ax.yaxis.set_label_text('FPOO Rate')
fig.set_size_inches(7,5)
plt.show()
```

```python
#Some people who start with opioid addictions are reported to transition to heroin use. 
#What is the relationship in Virginia counties between opioid overdoses and heroin overdoses? 
#Scatterplot diagrams show relationships between two variables. By just looking at this graph, you can tell
#that there is a positive correlation between opiod and heroin overdoses. 

fig, ax = plt.subplots()
ax.scatter(opiods['FPOO-Rate'], opiods['FPOO-Rate'], alpha=0.5, c='b')
fig.suptitle('Heroin Overdose Rate vs. Opiod Overdose Rate in VA')
ax.xaxis.set_label_text('Fentanyl and/or heroin overdoses per 100,000 county residents')
ax.yaxis.set_label_text('Opioid overdoses per 100,000 county residents')
fig.set_size_inches(7,5)
plt.show()
```

```python
uspres = pd.read_csv('presidents.csv')
```

```python
uspres
```

```python
#Which states are associated with the greatest number of United States presidents in terms of the presidents’ birthplaces?
#I used a very simple bar chart to show which state has the greatest number of US presidents. 

uspres['State'].value_counts().plot(kind='bar')
plt.xlabel("State")
plt.ylabel("Frequency")
plt.title("States Where US Presidents were Born", y=1.02);
```

```python
#How have vehicle sales in the United States varied over time?
#One of the key lessons in BUAD5052 is time series. This question is a perfect example 
#where you can show the trend over time using a time series in a line graph. 

fig, ax = plt.subplots()
plt.plot(tsv['DATE'],tsv['TOTALNSA'], alpha=0.5, c='b')
fig.suptitle('United States Vehicle Sales')
ax.xaxis.set_label_text('Date')
ax.yaxis.set_label_text('Total Sales for the Month')
fig.set_size_inches(7,5)
plt.show()
```
